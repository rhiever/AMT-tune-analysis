# AMT tunes data

Generated 3-tune ring tones and had AMT users rate them on scale of 1-6.

First tune is constant 0 (low tone). Next two tunes are varied for the experiment. Ring tones are named `(a, b)`, where `a` and `b` are tune values ranging from 0 (low tone) to 11 (high tone).

(13, 13) is a control with no sound at all.

Each AMT user rated 4 random tunes and 1 control tune.

User ratings are omitted if they:

* provide an age < 18

* do not properly mark the control tune as "no sound"

* provide all ratings in less than 10 seconds (which is impossible, as listening to all 5 tunes requires more than 10 seconds)

# Read in and look at summary data for the tunes data

```python
def read_tunes_data():
    tunes_data = read_csv("tune_results.csv", sep="\t")
    tunes_data = tunes_data[tunes_data["tuneA"] != 13]
    tunes_data = tunes_data[tunes_data["tuneB"] != 13]
    tunes_data = tunes_data[tunes_data["age"] < 80]
    tunes_data = tunes_data[tunes_data["age"] >= 18]
    return tunes_data
```

# Correlation between mean rating and variance in the ratings

The higher or lower a rating is, the less variance there is for that rating. The highest variance occurs at mean ratings of ~3.

This tells us that good tunes are more consistently liked, and bad tunes are more consistently disliked.

```python
from scipy.stats import pearsonr
from numpy import polyfit, poly1d
from pandas import read_csv

tunes_data = read_tunes_data()
tunes_data_grouped_tune = tunes_data.groupby(["tuneA", "tuneB"])

means = tunes_data_grouped_tune["rating"].mean()
variances = tunes_data_grouped_tune["rating"].var()
        
corr_coeff, pval = pearsonr(means, variances)

fit = polyfit(means, variances, 1)
fit_fn = poly1d(fit)

figure(figsize=(14, 6))
title("Pearson's r = %f,  p = %f | Linear fit: y = %f - %fx" % (corr_coeff, pval, round(fit[1], 3), round(abs(fit[0]), 3)), fontsize=16)
scatter(means, variances, s=25, color="green")
plot(means, fit_fn(means), "-k", lw=3)
xlabel("Mean", fontsize=16)
ylabel("Variance", fontsize=16)
xticks(fontsize=16)
yticks(fontsize=16)
;
```

## Piecewise fit to real ratings

Fit for only ratings <= 3, and then separately for only ratings >= 3. Does variance significantly go down at both ends of the ratings distribution?

```python
from scipy.stats import pearsonr
from numpy import polyfit, poly1d
from pandas import read_csv

tunes_data = read_tunes_data()
tunes_data_grouped_tune = tunes_data.groupby(["tuneA", "tuneB"])

means = tunes_data_grouped_tune["rating"].mean()
variances = tunes_data_grouped_tune["rating"].var()

sub_means = []
sub_variances = []

for i, mean in enumerate(means.values):
    if mean <= 3:
        sub_means.append(mean)
        sub_variances.append(variances.values[i])

means = array(sub_means)
variances = array(sub_variances)
        
corr_coeff, pval = pearsonr(means, variances)

fit = polyfit(means, variances, 1)
fit_fn = poly1d(fit)

figure(figsize=(14, 6))
title("Pearson's r = %f,  p = %f | Linear fit: y = %f + %fx" % (corr_coeff, pval, round(fit[1], 3), round(abs(fit[0]), 3)), fontsize=16)
scatter(means, variances, s=25, color="green")
plot(means, fit_fn(means), "-k", lw=3)
xlabel("Mean", fontsize=16)
ylabel("Variance", fontsize=16)
xticks(fontsize=16)
yticks(fontsize=16)
;
```

# How many ratings are required to capture the true rating?

Look at each tune's ratings. Randomly sample from the ratings until the tune's original distribution is captured. How many samples per tune are required?

## Assessment

It looks like each tune's distribution can be reasonably reconstructed with about 10 samples per tune. However, the sample means can range as much as $\pm1$ from the "true" mean tune rating, so that's far too small of a sample size.

If we use a sample size of ~20 user ratings, we get as close as $\pm0.5$ from the "true" mean tune rating, and ~30 user ratings get us as close as $\pm0.25$. ~40+ user ratings doesn't get us any more reliable of an estimate.

Note that with all sample sizes (10, 20, 30, 40), the individual tune rating distributions are not significantly different from each other.

From this analysis, I'd say the fewest ratings we'd want per tune would be about 20 to get a reasonable estimation of the rating. 10 if we don't care about a decent bit of noise in the data set.

```python
from pandas import read_csv
from scipy.stats import ranksums
from collections import defaultdict

tunes_data = read_tunes_data()
tunes_data_grouped_tune = tunes_data.groupby(["tuneA", "tuneB"])

ratings_needed_per_tune = defaultdict(int)
mean_ratings_per_tune = defaultdict(float)

for group in tunes_data_grouped_tune:
    tune = group[0]
    ratings = group[1]["rating"].values
    
    for sample_size in range(5, len(ratings) + 1):
        good_SS = True
        mean_ratings = []
        
        for repeat in range(100):
            sample_ratings = np.random.choice(ratings, size=sample_size, replace=False)
            pval = ranksums(ratings, sample_ratings)[1]
            if pval < 0.05:
                good_SS = False
                break
            mean_ratings.append(sample_ratings.mean())
                
        if good_SS:
            break
            
    ratings_needed_per_tune[tune] = (sample_size, len(ratings))
    mean_ratings_per_tune[tune] = (mean_ratings, ratings.mean())
```

## With ~10 samples

```python
tunes = []
needed = []
total = []

for i, tune in enumerate(sorted(ratings_needed_per_tune)):
    tunes.append(i)
    needed.append(ratings_needed_per_tune[tune][0])
    total.append(ratings_needed_per_tune[tune][1])
    
figure(figsize=(24, 6))
xlim(-1, 144)
ylim(0, 80)
ylabel("Number of ratings")
xlabel("Tune")
plot(tunes, needed, label="Needed to reconstruct full sample")
plot(tunes, total, label="Total samples")
legend()
```

```python
tunes = []
sampled = []
actual = []

for i, tune in enumerate(sorted(mean_ratings_per_tune)):
    for sample in mean_ratings_per_tune[tune][0]:
        tunes.append(i)
        sampled.append(sample)
        actual.append(mean_ratings_per_tune[tune][1])
    
figure(figsize=(24, 6))
xlim(-1, 144)
ylim(0, 6)
plot(tunes, sampled, "x")
plot(tunes, actual, "ro")
```

```python
tunes = []
needed = []
total = []

for i, tune in enumerate(sorted(ratings_needed_per_tune)):
    tunes.append(i)
    needed.append(ratings_needed_per_tune[tune][0])
    total.append(ratings_needed_per_tune[tune][1])
    
figure(figsize=(24, 6))
xlim(-1, 144)
ylim(0, 80)
ylabel("Number of ratings")
xlabel("Tune")
plot(tunes, needed, label="Needed to reconstruct full sample")
plot(tunes, total, label="Total samples")
legend()
```

```python
tunes = []
sampled = []
actual = []

for i, tune in enumerate(sorted(mean_ratings_per_tune)):
    for sample in mean_ratings_per_tune[tune][0]:
        tunes.append(i)
        sampled.append(sample)
        actual.append(mean_ratings_per_tune[tune][1])
    
figure(figsize=(24, 6))
xlim(-1, 144)
ylim(0, 6)
plot(tunes, sampled, "x")
plot(tunes, actual, "ro")
```

# Simulate random ratings for each tune. Does same correlation between mean rating and variance in rating show up?

Simulate 60 random samples for each tune and look at correlation between mean rating and variance.

No significant trend for random ratings. Thus the correlation between below- and above-average ratings and lower variance in those ratings is non-random.

```python
from collections import defaultdict

# simulate the data
tune_ratings = defaultdict(list)

for tuneA in range(12):
    for tuneB in range(12):
        samples = []
        for sample in range(60):
            samples.append(np.random.randint(0, 7))
        tune_ratings[(tuneA, tuneB)] = array(samples)
```

```python
from scipy.stats import pearsonr
from numpy import polyfit, poly1d

# fit the data
means = []
variances = []

for tune in sorted(tune_ratings):
    means.append(tune_ratings[tune].mean())
    variances.append(tune_ratings[tune].var())
    
means = array(means)
variances = array(variances)
        
corr_coeff, pval = pearsonr(means, variances)

fit = polyfit(means, variances, 1)
fit_fn = poly1d(fit)

figure(figsize=(14, 6))
title("Pearson's r = %f,  p = %f | Linear fit: y = %f + %fx" % (corr_coeff, pval, round(fit[1], 3), round(abs(fit[0]), 3)), fontsize=16)
scatter(means, variances, s=25, color="green")
plot(means, fit_fn(means), "-k", lw=3)
xlabel("Mean", fontsize=16)
ylabel("Variance", fontsize=16)
xticks(fontsize=16)
yticks(fontsize=16)
;
```
