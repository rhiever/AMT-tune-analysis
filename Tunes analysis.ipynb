# AMT tunes data

Generated 3-tune ring tones and had AMT users rate them on scale of 1-6.

First tune is constant 0 (low tone). Next two tunes are varied for the experiment. Ring tones are named `(a, b)`, where `a` and `b` are tune values ranging from 0 (low tone) to 11 (high tone).

(13, 13) is a control with no sound at all.

Each AMT user rated 4 random tunes and 1 control tune.

User ratings are omitted if they:

* provide an age < 18

* do not properly mark the control tune as "no sound"

* provide all ratings in less than 10 seconds (which is impossible, as listening to all 5 tunes requires more than 10 seconds)

# Read in and look at summary data for the tunes data

```python
def read_tunes_data():
    tunes_data = read_csv("tune_results.csv", sep="\t")
    tunes_data = tunes_data[tunes_data["tuneA"] != 13]
    tunes_data = tunes_data[tunes_data["tuneB"] != 13]
    tunes_data = tunes_data[tunes_data["age"] < 80]
    tunes_data = tunes_data[tunes_data["age"] >= 18]
    return tunes_data
```

# Heatmap of mean rating and variance in ratings

y-axis is the first value of the tune, x-axis is the second value of the tune. So tune (0, 0) is on the top left; tune (11, 11) is on the bottom right.

There seems to be a correlation between high mean rating and lower variance in that rating. Let's explore that more...

```python
from pandas import read_csv

tunes_data = read_tunes_data()
tunes_data_grouped_tune = tunes_data.groupby(["tuneA", "tuneB"])

means = tunes_data_grouped_tune["rating"].mean().reshape((12, 12))
variances = tunes_data_grouped_tune["rating"].var().reshape((12, 12))

figure(figsize=(12, 6))
subplot(121)
imshow(means, interpolation="nearest")
colorbar(shrink=0.8)
title("Mean rating", fontsize=16)
xticks(range(12), fontsize=14)
yticks(range(12), fontsize=14)

subplot(122)
imshow(variances, interpolation="nearest")
colorbar(shrink=0.8)
title("Variance in ratings", fontsize=16)
xticks(range(12), fontsize=14)
yticks(range(12), fontsize=14)
tight_layout();
```

# "Spaghetti sauce" experiment

The bulk of the ratings sit at 3.0, which is the average.

Could it be that there's a split of opinion on several of these? Some people love it, others hate it, and because of that the average comes out in the middle ground.

From the data, there doesn't seem to be a huge split in opinion, e.g., people regularly voting 0 and others voting 6.

In fact, most raters seem to avoid voting on the extremes except on really bad tunes. For the most part, raters seem to vote 2 and 4 to indicate whether they like it or not.

# Negative correlation between mean rating and variance in the ratings

The higher a rating is, the less variance there is for that rating.

This tells us that good tunes are more consistently liked -- at least moreso than bad tunes.

Alternatively, the other way to look at it is that bad tunes are less consistently disliked.

```python
from scipy.stats import pearsonr
from numpy import polyfit, poly1d
from pandas import read_csv

tunes_data = read_tunes_data()
tunes_data_grouped_tune = tunes_data.groupby(["tuneA", "tuneB"])

means = tunes_data_grouped_tune["rating"].mean()
variances = tunes_data_grouped_tune["rating"].var()

corr_coeff, pval = pearsonr(means, variances)

fit = polyfit(means, variances, 1)
fit_fn = poly1d(fit)

figure(figsize=(14,6))
title("Pearson's r = %f,  p = %f | Linear fit: y = %f - %fx" % (corr_coeff, pval, round(fit[1], 3), round(abs(fit[0]), 3)), fontsize=16)
scatter(means, variances, s=25, color="green")
plot(means, fit_fn(means), "-k", lw=3)
xlabel("Mean", fontsize=16)
ylabel("Variance", fontsize=16)
xticks(fontsize=16)
yticks(fontsize=16)
;
```

# Heatmap of ratings by gender

There's some pretty clear tune preferences by gender. Opinion is split on different tunes as well.

This has some interesting implications, at least for subjective IEC design like this. It means that even a basic difference like gender can cause raters to rate items differently.

Need to run stats on these difference to confirm this qualitative observation.

```python
from pandas import read_csv

tunes_data = read_tunes_data()

tunes_data_males = tunes_data[tunes_data["gender"] == "male"]
tunes_data_males_grouped_tune = tunes_data_males.groupby(["tuneA", "tuneB"])
males_means = tunes_data_males_grouped_tune["rating"].mean().reshape((12, 12))
males_variances = tunes_data_males_grouped_tune["rating"].var().reshape((12, 12))

tunes_data_females = tunes_data[tunes_data["gender"] == "female"]
tunes_data_females_grouped_tune = tunes_data_females.groupby(["tuneA", "tuneB"])
females_means = tunes_data_females_grouped_tune["rating"].mean().reshape((12, 12))
females_variances = tunes_data_females_grouped_tune["rating"].var().reshape((12, 12))

figure(figsize=(16, 16))
subplot(221)
imshow(males_means, interpolation="nearest")
colorbar(shrink=0.8)
title("Mean rating (males)", fontsize=16)
xticks(range(12), fontsize=14)
yticks(range(12), fontsize=14)

subplot(222)
imshow(females_means, interpolation="nearest")
colorbar(shrink=0.8)
title("Mean rating (females)", fontsize=16)
xticks(range(12), fontsize=14)
yticks(range(12), fontsize=14)

subplot(223)
imshow(males_variances, interpolation="nearest", vmin=1.2, vmax=4.5)
colorbar(shrink=0.8)
title("Variance in ratings (males)", fontsize=16)
xticks(range(12), fontsize=14)
yticks(range(12), fontsize=14)

subplot(224)
imshow(females_variances, interpolation="nearest", vmin=1.2, vmax=4.5)
colorbar(shrink=0.8)
title("Variance in ratings (females)", fontsize=16)
xticks(range(12), fontsize=14)
yticks(range(12), fontsize=14)

tight_layout();
```
